**23.** Given a DataFrame of numeric values, say
```python
df = pd.DataFrame(np.random.random(size=(5, 3))) # a 5x3 frame of float values
```

how do you subtract the row mean from each element in the row?

````python
df = pd.DataFrame(np.random.random(size=(5, 3)))
print("The current DataFrame\n", df)
print("\nThe result is:")
# I substract for each column(axis = 1) the mean of each row(axis = 0)
df.sub(df.mean(axis = 1), axis = 0)
````

**24.** Suppose you have DataFrame with 10 columns of real numbers, for example:

```python
df = pd.DataFrame(np.random.random(size=(5, 10)), columns=list('abcdefghij'))
```
Which column of numbers has the smallest sum? (Find that column's label.)

````python
df = pd.DataFrame(np.random.random(size=(5, 10)), columns=list('abcdefghij'))

# I like to reaffirm so that's why I printed the sum of each column
print(df.sum(), "\n")
# The just using the method idxmin() after the sum(), I find the index of the minimum value
print("The column ", df.sum().idxmin(), " has a sum of ", df.sum().min())
````
