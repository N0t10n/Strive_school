**72. How to swap two rows of an array?** 

```python
# n is the length of the array
def row_swaper(n):
    x = np.random.randint(n, size = (2, n))
    print(f"The generated array is:\n{x}")
    print("\n")
    # Here's where comes the swapping magic
    x[[0, 1]] = x[[1, 0]]
    return f"The swapped array is:\n{x}"

print(row_swaper(10))
```

**75. How to compute averages using a sliding window over an array?**

```python
def moving_average(len_x, len_y):
    x = np.random.randint(len_x)
    y = np.random.randint(len_y)
    print(x, y)
    # valid is given for overlaping elements
    return np.convolve(x, y, 'valid')

print(moving_average(10, 10))
```
