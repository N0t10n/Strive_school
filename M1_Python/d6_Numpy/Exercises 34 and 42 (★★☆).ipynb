**42. Consider two random array A and B, check if they are equal**

```python
def comp_2_arr(num, size):
    A = np.random.randint(num, size = size)
    B = np.random.randint(num, size = size)
    print(
        f"Array A = {A}\nArray B = {B}"
    )
    return A == B

print(comp_2_arr(5, 5))
```
